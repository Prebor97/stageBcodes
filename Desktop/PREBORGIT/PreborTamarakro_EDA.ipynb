```python
from sklearn.model_selection import train_test_split
# i want to make regression analysis on only south africa and india
regforeignEx_data_1 = foreignEx_data_1[['SOUTH AFRICA - RAND/US$', 'INDIA - INDIAN RUPEE/US$']]
```

```python
rate_target = regforeignEx_data_1 ['INDIA - INDIAN RUPEE/US$']
```

```python
x_train,x_test, y_train, y_test = train_test_split(regforeignEx_data_1,rate_target, test_size = 0.3, random_state = 42)
```

```python
# fit the linear model to the training data
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
predi_value = linear_model.predict(x_test)
```

```python
print(predi_value)# I wanted to see if there was a way to visualize the predicted value and actual value, then i realise the training set and testing set are randomly chosen by python, so there was no way to make comparison in that regard. Rather i just used the a metrics(mse). But i would have to visualize both predicted and actual value.
```

```python
from sklearn.metrics import mean_squared_error  
mse = mean_squared_error(y_test, predi_value)
```
